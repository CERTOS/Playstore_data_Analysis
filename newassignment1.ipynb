```python
#importing things
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
```

```python
##reading files
play_store_data=pd.read_csv('playstore analysis.csv')
```

```python
#checking the impoted file
play_store_data.head()
```

```python
#checking the null values in the csv file system
play_store_data.isnull().sum()
```

```python
#dropping the null values present in rating column
play_store_data.dropna(how='any', subset=['Rating'], axis=0, inplace = True)
play_store_data.Rating.isnull().sum()
```

```python
#checking the null value availabity in Anderiod Ver column
play_store_data.loc[play_store_data['Android Ver'].isnull()]
```

```python
#dropping the 3rd null record of the anderiod ver column
play_store_data.drop([10472], inplace = True)
```

```python
#checking the result of the dropping and confirming the drop
play_store_data.loc[play_store_data['Android Ver'].isnull()]
```

```python
#replacing others with mode
play_store_data['Android Ver'].fillna(play_store_data['Android Ver'].mode()[0], inplace=True)
```

```python
#replacing the null values of Current ver with the most frequent i.e mode of the data
play_store_data['Current Ver'].fillna(play_store_data['Current Ver'].mode()[0], inplace=True)
```

```python
#checking data types of the columns 
dns = pd.DataFrame(play_store_data)
print (dns.dtypes)
```

```python
print("mine analysis is : Reviews and installs need to be brought to numeric types")
```

```python
#removing $ from the price column and converting to float
price = []
for i in play_store_data['Price']:
    if i[0]=='$':
        price.append(i[1:])
    else:
        price.append(i)  
```

```python
play_store_data.drop(labels=play_store_data[play_store_data['Price']=='Everyone'].index, inplace = True)
play_store_data['Price']= price
play_store_data['Price']= play_store_data['Price'].astype('float')
```

```python
#confirming the changes
print(play_store_data["Price"].dtypes) 
```

```python
#removing the , and + from installs column
install = []
for j in play_store_data['Installs']:
    install.append(j.replace(',','').replace('+','').strip())

play_store_data['Installs']= install
play_store_data['Installs']= play_store_data['Installs'].astype('int')
```

```python
#converting it to integer
play_store_data['Reviews']= play_store_data['Reviews'].astype('int')
```

```python
#confirming the changes
print(play_store_data["Installs"].dtypes) 
```

```python
#checking the ratings of the apps which are more than 5 or less than 1
play_store_data.loc[play_store_data.Rating < 1] & play_store_data.loc[play_store_data.Rating > 5]
```

```python
##since the set is empty it is taken in condieration that there is no record with rating less than 1 or more than 5
```

```python
##checking the records which has reviews more than installs
play_store_data.loc[play_store_data['Reviews'] > play_store_data['Installs']]
```

```python
#dropping those records
temp = play_store_data[play_store_data['Reviews']>play_store_data['Installs']].index
play_store_data.drop(labels=temp, inplace=True)
```

```python
#checking again for confirmation
play_store_data.loc[play_store_data['Reviews'] > play_store_data['Installs']]
```
